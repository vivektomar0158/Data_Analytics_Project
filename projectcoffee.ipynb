```python
revenue = df.groupby(['coffee_name']).sum(['money']).reset_index().sort_values(by = 'money', ascending = False)
revenue
```

```python
plt.figure
ax = sns.barplot(data = revenue, x = 'money', y = 'coffee_name', color ='steelblue')
ax.bar_label(ax.containers[0], fontsize=6)
plt.xlabel('Revenue')
```

```python
month_sale =df.groupby(['coffee_name','month']).count()['date'].reset_index().rename(columns ={'date':'count'}).pivot(index='month',columns='coffee_name',values='count').reset_index()
month_sale
```

```python
month_sale.describe().T.loc[:,['min','max']]
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=month_sale)
plt.legend(loc='upper left')
plt.xticks(range(len(month_sale['month'])),month_sale['month'],size='small')
```

```python
weekday_sales = df.groupby(['day']).count()['date'].reset_index().rename(columns={'date':'count'})
weekday_sales
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=weekday_sales,x='day',y='count',color='steelblue')
plt.xticks(range(len(weekday_sales['day'])),['Sun','Mon','Tue','Wed','Thur','Fri','Sat'],size='small')
```

```python
daily_sales = df.groupby(['coffee_name','date']).count()['datetime'].reset_index().rename(columns = {'datetime':'count'}).pivot(index = 'date', columns = 'coffee_name',values ='count').reset_index().fillna(0)
daily_sales
```

```python
daily_sales.iloc[:,1:].describe().T.loc[:,['min','max']]
```

```python
hourly_sales =df.groupby(['hours']).count()['date'].reset_index().rename(columns={'date':'count'})
hourly_sales
```

```python
sns.barplot(data=hourly_sales,x='hours',y='count',color='steelblue')
```

```python
hourly_sales_by_coffee = df.groupby(['hours','coffee_name']).count()['date'].reset_index().rename(columns={'date':'count'}).pivot(index='hours',columns='coffee_name',values='count').fillna(0).reset_index()
hourly_sales_by_coffee
```

```python
fig, axs = plt.subplots(2, 4, figsize=(20, 10))
# Flatten the array of subplots for easy iteration
axs = axs.flatten()
# Loop through each column in the DataFrame, skipping the 'Index' column
for i, column in enumerate(hourly_sales_by_coffee.columns[1:]): # Skip the firstcolumn ('Index')
    axs[i].bar(hourly_sales_by_coffee['hours'], hourly_sales_by_coffee[column])
    axs[i].set_title(f'{column}')
    axs[i].set_xlabel('hours')
#axs[i].set_ylabel('Sales')
    plt.tight_layout()
# Show the plot
plt.show
```
